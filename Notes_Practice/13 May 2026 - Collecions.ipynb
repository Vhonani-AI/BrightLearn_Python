```python
#list
students = ["John", "Sarah", "Mike", "John", "Lisa"]

#Dictionary
marks = {
    "John": 75,
    "Sarah": 88,
    "Mike": 67,
    "Lisa": 90
    }
```

Remove duplicates from students

```python

students.remove("John")
print(students)
```

```python
#removing duplicates on a long list
set_students = set (students)
print(set_students)
```

Add a new student called "David"

```python
students.append("David")
print(students)
```

Print Sarah’s mark

```python
print(marks["Sarah"])
```

Update Mike’s mark to 80

```python
marks["Mike"] = 80
print(marks)
```
