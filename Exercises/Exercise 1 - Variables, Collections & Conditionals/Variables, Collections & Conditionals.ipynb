```python
students = {
        "student_name": "Amara",
        "age": 19,
        "subject": "Math",
        "score": 72,
        "passed": True
    }
```

```python
#3.1
print(students["score"] > 60)
print(students["age"] == 20)
print(students["subject"] != "Science")
print(students["score"] >= 72)
```

```python
#3.2
print(students["score"] > 60 and students["passed"] == True)
print(students["age"] < 18 or students["score"] > 70)
print(not students["passed"])
print(students["score"] > 80 and students["subject"] == "Math")
```

```python
#3.3
if students["score"] >= 75:
    print("Distinction")
elif students["score"] >= 50:
    print("Pass")
else:
    print("Fail")
```

```python
#3.4
if students["score"] > 60 and students["subject"] == "Math":
    print("Math high achiever")
else:
    print("Does not meet criteria")
```
